Since the Avengers were created in the 1960's, the values in the "Year" column showing up on the histogram in the bin for 1900 to 1910 seem erroneous. Let's clean that up. We only want to keep the Avengers who were introduced after 1960.

```python
true_avengers = avengers[avengers["Year"]>=1960]
true_avengers["Year"].hist()
```

```python
av_deaths = ["Death"+ str(i) for i in range(1,6)]
true_avengers[av_deaths]
```

As can be seen above, there are five fields (Death1 to Death5), each of which contains a binary value representing whether a superhero experienced that death or not. Let's combine that information in a single field so we can perform numerical analysis on it more easily.

```python
true_avengers["Deaths"] = true_avengers[av_deaths].replace({"YES":1,"NO":0}).sum(axis=1)
```

```python
true_avengers["Deaths"].astype(int)
```

```python
true_avengers["Deaths"].plot(kind="hist")
```

Lastly, let's verify that the Years since joining field accurately reflects the Year column.

```python
joined_accuracy_count  = int()
joined_accuracy_count = (true_avengers['Years since joining'] == (2015 - true_avengers['Year'])).sum()
print(joined_accuracy_count)
```

Let's see if we can use the data we cleaned to plot the number of deaths of the avengers by gender

```python
import seaborn as sns
g = sns.FacetGrid(true_avengers,col="Gender")
g.map(sns.distplot,"Deaths")
```
